##Perform Data Preprocessing and Analysis on Carbon_Emmissions dataset

###Import all libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

###Reading the data

###Visualising the data

```python
plt.scatter(x='Engine Size(L)',y='CO2 Emissions(g/km)',data=df,color='blue',label='scatter')
plt.xlabel('Engine Size(L)')
plt.ylabel('CO2 Emissions(g/km)')
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(8,8),dpi=50)
sns.pairplot(df)
```

```python
plt.figure(figsize=(8,8),dpi=90)
sns.boxplot(x='Engine Size(L)',y='CO2 Emissions(g/km)',data=df)
plt.xticks(rotation=90)
plt.title('CO2 Emissions based on Engine size')
```

```python
plt.figure(figsize=(8,8),dpi=70)
sns.barplot(x='Make',y='CO2 Emissions(g/km)',hue='Cylinders',data=df)
plt.xticks(rotation=90)
```

```python
pd.unique(df['Fuel Type'])
```

```python
plt.figure(figsize=(9,9),dpi=70)
sns.violinplot(x='Fuel Type',y='CO2 Emissions(g/km)',data=df)
```

```python
plt.figure(figsize=(9,9),dpi=70)
df1=df.corr()
sns.heatmap(df1,annot=True)
```

```python
plt.figure(figsize=(9,9),dpi=70)
sns.histplot(data=df,x='Vehicle Class',y='CO2 Emissions(g/km)',color='#32a88b')
plt.xticks(rotation=90)
plt.grid()
```

```python
plt.figure(figsize=(70,70),dpi=90)
sns.catplot(x='Vehicle Class', y='CO2 Emissions(g/km)', hue='Fuel Type',kind='bar', data=df);
plt.xticks(rotation=90)
plt.title('Vechile class in CO2 Emissions containing Fuel type')
```
